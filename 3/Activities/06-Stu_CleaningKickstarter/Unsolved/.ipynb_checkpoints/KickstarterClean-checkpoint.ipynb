```python
import pandas as pd
```

```python
# The path to our CSV file
file = "Resources/KickstarterData.csv"
# Read our Kickstarter data into pandas
df= pd.read_csv(file)
df.head()
```

```python
# Get a list of all of our columns for easy reference
df.columns
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
my_kickstarter_df = df.loc[:, ["name", "goal", "pledged",
                                    "state", "country", "staff_pick", "backers_count", "spotlight"]]
my_kickstarter_df
```

```python
# Remove projects that made no money at all
my_kickstarter_df = my_kickstarter_df.loc[(
    my_kickstarter_df["pledged"] > 0)]
my_kickstarter_df.head()
```

```python
# Collect only those projects that were hosted in the US

# Create a list of the columns for the columns you want to report
report_columns = ["name", "goal", "pledged",
                   "state", "country", "staff_pick", 
                   "backers_count", "spotlight"]
# Create a new df for "US" with the columns above, use loc as this is a filter, so find country and set to US. 
us_kickstarter_df = my_kickstarter_df.loc[my_kickstarter_df["country"]=="US", report_columns]
us_kickstarter_df.head()
```

```python
#I have made the overall average for all pledged
average_pledge_overall = round(us_kickstarter_df["pledged"].mean(),2)
print (average_pledge_overall)
```

```python
# Create a new column that finds the average amount pledged to a project, this is how much by many backers 
average_pledge_per_project = us_kickstarter_df["pledged"]/us_kickstarter_df["backers_count"]
us_kickstarter_df
average_pledge_per_project
us_kickstarter_df["Average_Pledge"]=average_pledge_per_project
us_kickstarter_df.head()
```

```python
# First convert "average_donation", "goal", and "pledged" columns to float
# Then Format to go to two decimal places, include a dollar sign, and use comma notation

us_kickstarter_df["Average_Pledge"] = us_kickstarter_df["Average_Pledge"].astype(float).map(
    "${:,.2f}".format)
us_kickstarter_df["goal"] = us_kickstarter_df["goal"].astype(float).map("${:,.2f}".format)
us_kickstarter_df["pledged"] = us_kickstarter_df["pledged"].astype(float).map("${:,.2f}".format)

us_kickstarter_df.head()
```
