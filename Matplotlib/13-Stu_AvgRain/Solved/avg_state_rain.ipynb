```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
rain_df = pd.read_csv("../Resources/avg_rain_aus.csv")
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rain_df))
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(10,5))
plt.bar(x_axis, rain_df["Milimeters"], color='r', alpha=0.5)
plt.xticks(x_axis, rain_df["City"], rotation="vertical")
plt.tight_layout()
```

```python
# Set x and y limits
plt.xlim(-1, len(x_axis))
plt.ylim(0, max(rain_df["Milimeters"])+100)
```

```python
# Set a Title and labels
plt.title("Average Rain per City")
plt.xlabel("City")
plt.ylabel("Average Amount of Rainfall in Milimeters")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("../Images/avg_state_rain.png")
plt.show()
```
