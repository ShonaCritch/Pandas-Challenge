```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x_axis_data = np.arange(1,13,1)
x_axis_data
```

```python
# Average weather temp
points = [3.92, 5.6, 10.64, 16.8, 22.4, 28.0, 30.24, 29.12, 25.2, 18.48, 12.88, 6.72]
```

```python
# Plot the line
plt.plot(x_axis_data, points)
plt.show()
```

```python
# Convert to Fahrenheit F = (C/0.56)+32
points_F = [(x/0.56) + 32 for x in points]
points_F
```

```python
# Plot using Fahrenheit
plt.plot(x_axis_data, points_F)
plt.show()
```

```python
# Plot both on the same chart
plt.plot(x_axis_data, points)
plt.plot(x_axis_data, points_F)
plt.show()
```
